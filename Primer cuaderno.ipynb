```python
calificaciones = []
calif=int(input("¿Cuántas calificaciones deseas agregar?: "))
for i in range(calif) :
  contador = False
  while contador ==False:
    ca = int(input(f"Dame la calificación {i+1} : "))
    if ca > 10 or ca < 0:
      print("LA CALIFICACION ES ERRONEA")
    else :
      calificaciones.append(ca)
      contador = True
print(calificaciones)
cf = sum(calificaciones)/calif
if cf <= 5 :
  print("REPROBÓ")
elif cf > 5 :
  print("APROBÓ")
print(f"PROMEDIO : {cf}")



```

```python
def multiplicacion (lista):
  try:
    for i in range(len(lista)):
      lista_mult [i+1] = lista_mult[i] * lista_mult[i+1]
      print(lista_mult)
    resultado = lista_mult[-1]
  except IndexError :
    print("SE ESTÁ MULTIPLICANDO")
  return 0
```

```python
def promedio (lista):
  rs = sum (lista)
  na = len (lista)
  resultado = rs / na
  return resultado
```

```python
lista_sum = []
lista_neg = []
lista_res = []
lista_prom = []
lista_mult = [1]
print("CALCULADORA")
print("1 = SUMA")
print("2 = SUMA DE NÚMEROS NEGATIVOS")
print("3 = RESTA")
print("4 = PROMEDIO")
print("5 = MULTIPLICACIÓN")
print("6 = DIVISIÓN")
op=int(input("¿Qué operación desea realizar?"))
if op == 1 :
  cantidadsum = int(input("¿Cuántos números deseas sumar?"))
  for i in range (cantidadsum):
   a = int(input(f"Dame el número {i+1} : "))
   lista_sum.append(a)
   print(lista_sum)
  rssum = sumar(lista_sum)
  print(f"EL RESULTADO ES : {rssum}")
elif op == 2 :
  cantidadneg = int(input("¿Cuántos números deseas restar?"))
  for i in range (cantidadneg) :
    b = int(input(f"Dame el número {i+1}?"))
    lista_neg.append(b)
    print(lista_neg)
  rsneg = negativo(lista_neg)
  print(f"EL RESULTADO ES : {rsneg}")
elif op == 3 :
  cantidadres = int(input("¿Cuántos números deseas restar?"))
  num1 = int(input("Dame el número 1 :"))
  for i in range (1,cantidadres):
   c = int(input(f"Dame el número {i+1} : "))
   lista_res.append(c)
   print(lista_res)
  rsres = restar(lista_res) + num1
  print(f"EL RESULTADO ES : {rsres}")
elif op == 4 :
  cantidadprom = int(input("¿Cuántos números deseas sacar promedio?"))
  num1 = int(input("Dame el número 1 :"))
  for i in range (1,cantidadprom):
   d = int(input(f"Dame el número {i+1} : "))
   lista_prom.append(d)
   print(lista_prom)
  rsprom = promedio(lista_prom) + num1
  print(f"EL RESULTADO ES : {rsprom}")
elif op == 5 :
  cantidadmult=int(input("¿Cuántos números deseas multiplicar?"))
  for i in range (cantidadmult):
    e = int(input(f"Dame el número {i+1} : "))
    lista_mult.append(e)
    print(lista_mult)
  multiplicacion(lista_mult)
  rsmult=lista_mult[-1]
  print(f"EL RESULTADO ES : {rsmult}")
elif op == 6 :
  rsdiv1 = int(input("Dame el número 1 :"))
  rsdiv2 = int(input("Dame el número 2 :"))
  resultadosdiv=rsdiv1 * rsdiv2
  print(f"EL RESULTADO ES : {resultadosdiv}")
```
